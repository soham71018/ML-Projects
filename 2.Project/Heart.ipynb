# Data preprocessing and cleaning

```python
df_clean=  pd.get_dummies(df,drop_first=True)
df_clean = df_clean.astype(int)
df_clean
```

```python
#applying standard scaler
from sklearn.preprocessing import StandardScaler
num_cols = ['Age', 'RestingBP', 'Cholesterol', 'MaxHR', 'Oldpeak']
scaler = StandardScaler()
df_clean[num_cols] = scaler.fit_transform(df_clean[num_cols])
df_clean.head()
```

# Feature Engineering and Extraction

```python
from scipy.stats import pearsonr
import pandas as pd

# Define features for heart disease project
selected_features = [
    'FastingBS', 'Sex_M',
    'ChestPainType_ATA', 'ChestPainType_NAP', 'ChestPainType_TA',
    'RestingECG_Normal', 'RestingECG_ST',
    'ExerciseAngina_Y',
    'ST_Slope_Flat', 'ST_Slope_Up'
]

# Calculate Pearson correlation with target variable (HeartDisease)
correlations = {
    feature: pearsonr(df_clean[feature], df_clean['HeartDisease'])[0] 
    for feature in selected_features
}
correlation_df = pd.DataFrame(list(correlations.items()), columns=['Feature', 'Pearson Correlation'])
correlation_df = correlation_df.sort_values(by='Pearson Correlation', ascending=False)
print(correlation_df)
```

```python
from scipy.stats import chi2_contingency 
import pandas as pd 

alpha = 0.05 

chi_features = [
    'FastingBS', 'Sex_M',
    'ChestPainType_ATA', 'ChestPainType_NAP', 'ChestPainType_TA',
    'RestingECG_Normal', 'RestingECG_ST',
    'ExerciseAngina_Y',
    'ST_Slope_Flat', 'ST_Slope_Up'
]

chi2_results = {}
for col in chi_features:
    contingency = pd.crosstab(df_clean[col], df_clean['HeartDisease'])
    chi2_stat, p_val, _, _ = chi2_contingency(contingency)
    decision = 'Reject Null (Keep Feature)' if p_val < alpha else 'Accept Null (Drop Feature)'
    chi2_results[col] = {
        'chi2_statistic': chi2_stat,
        'p_value': p_val,
        'Decision': decision
    }

chi2_df = pd.DataFrame(chi2_results).T
chi2_df = chi2_df.sort_values(by='p_value')
print(chi2_df)
```
